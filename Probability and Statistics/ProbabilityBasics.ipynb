Probability is simply how likely something is to happen.

Probability is a branch of mathematics that deals with the study of uncertainty and randomness. It provides a framework for quantifying the likelihood of different outcomes in uncertain situations. Here’s a basic explanation of the core concepts:

 1. Sample Space (S)
- The sample space is the set of all possible outcomes of a random experiment.
- Example: If you flip a coin, the sample space is \( S = \{ \text{Heads}, \text{Tails} \} \).

2. Event
- An event is a subset of the sample space, representing one or more outcomes.
- Example: In the coin flip, an event could be getting a head, so \( A = \{ \text{Heads} \} \).

3. Probability of an Event (P)
- Probability is a measure of how likely an event is to occur. It’s a number between 0 and 1, where:
  - 0 means the event will not happen.
  - 1 means the event will definitely happen.
  - A probability close to 0 means the event is unlikely, while a probability close to 1 means it is likely.
- Example: The probability of getting heads in a fair coin flip is \( P(\text{Heads}) = 0.5 \).

4. Axioms of Probability
Probability follows three basic rules or axioms:
1. Non-Negativity: The probability of any event \( A \) is always non-negative, \( P(A) \geq 0 \).
2. Normalization: The probability of the entire sample space is 1, \( P(S) = 1 \).
3. Additivity: If two events \( A \) and \( B \) cannot both happen at the same time (they are mutually exclusive), then the probability of either \( A \) or \( B \) happening is the sum of their probabilities, \( P(A \cup B) = P(A) + P(B) \).

5. Complementary Events
- The complement of an event \( A \) (denoted as \( A^c \)) is the event that \( A \) does not happen.
- The probability of the complement is \( P(A^c) = 1 - P(A) \).
- Example: If the probability of raining tomorrow is \( P(\text{Rain}) = 0.3 \), the probability that it won’t rain is \( P(\text{No Rain}) = 1 - 0.3 = 0.7 \).

6. Union and Intersection of Events
- Union: The event that at least one of \( A \) or \( B \) occurs is represented by \( A \cup B \). The probability of \( A \cup B \) is found by adding their probabilities and subtracting the probability that both happen together: 
  \[  P(A \cup B) = P(A) + P(B) - P(A \cap B) \]
- Intersection: The event that both \( A \) and \( B \) occur is represented by \( A \cap B \). The probability of \( A \cap B \) is the probability that both events happen at the same time.

7. Conditional Probability
- Conditional probability is the probability of an event \( A \) given that another event \( B \) has occurred, denoted \( P(A|B) \). It’s calculated using:
  \[  P(A|B) = \frac{P(A \cap B)}{P(B)}  \]
  provided \( P(B) > 0 \).

8. Independence
- Two events \( A \) and \( B \) are independent if the occurrence of one does not affect the probability of the other. Mathematically, \( A \) and \( B \) are independent if:
  \[  P(A \cap B) = P(A)P(B)  \]

9. Bayes’ Theorem
- Bayes’ Theorem helps in updating probabilities based on new evidence. It’s particularly useful in situations where you want to reverse conditional probabilities:
  \[  P(A|B) = \frac{P(B|A)P(A)}{P(B)}  \]

Summary
Probability helps us quantify and reason about uncertainty. By understanding the basics of probability, you can make informed predictions and decisions in situations where outcomes are uncertain.


 1. What is Probability?
- Probability is a way of guessing how likely something is to happen.
- Imagine you have a box of chocolates. Some are your favorite, and some are not. If you reach into the box without looking, probability helps you figure out how likely it is you'll grab your favorite chocolate.

 2. Sample Space (All Possible Outcomes)
- The sample space is just a fancy way of saying “all the things that could happen.”
- Example: If you flip a coin, there are two things that could happen: you get heads, or you get tails. So, the sample space is "heads" and "tails."

 3. Event (What You Care About)
- An event is something specific you’re interested in.
- Example: If you only care about getting heads when you flip the coin, then “getting heads” is your event.

 4. How Likely Is It? (Probability)
- Probability tells you how likely it is that your event will happen. It’s a number between 0 and 1:
  - If the probability is 0, it means it won’t happen.
  - If it’s 1, it means it will definitely happen.
  - If it’s somewhere in between, it tells you how likely it is.

 5. Examples:
- Coin Flip: If you flip a coin, there's a 50% chance it will land on heads and a 50% chance it will land on tails. So, the probability of getting heads is 0.5, and the same for tails.
  
- Rolling a Die: If you roll a die (a cube with numbers 1 to 6), there’s a 1 in 6 chance you’ll get any particular number, like a 3. That means the probability of rolling a 3 is \( \frac{1}{6} \).

 6. What About if It Doesn’t Happen?
- If you know the chance of something happening, you can figure out the chance of it not happening. 
- Example: If there’s a 70% chance it will rain tomorrow, there’s a 30% chance it won’t rain. 

 7. Adding Chances Together
- If you want to know the chance of one thing OR another thing happening, you can sometimes add the probabilities.
- Example: If there’s a 10% chance you’ll have ice cream today and a 20% chance you’ll have cake, the chance of having ice cream OR cake is 10% + 20% = 30%.

 8. What if One Thing Depends on Another?
- Sometimes, one thing happening depends on something else. This is called conditional probability.
- Example: If it’s cloudy, the chance of rain might be higher because rain often follows clouds.

 9. Independence
- Some events don’t affect each other at all. Like flipping a coin twice—the first flip doesn’t change the second flip. These are independent events.

 10. Bayes' Theorem (Advanced but Fun!)
- This is like being a detective. If you get new clues, you can change how you think about what might happen.

Summary:
Probability is like a fun guessing game where you use numbers to figure out how likely things are to happen. Whether it’s picking a chocolate, rolling a die, or guessing the weather, probability helps you make smart guesses!

```python
#calculate probability for rolling a die
#probability of rolling a 4 on a fair six-sided die?
import numpy as np

# Problem 1: Probability of getting heads in a coin flip
def coin_flip_probability():
    outcomes = ['Heads', 'Tails']
    probability_of_heads = outcomes.count('Heads') / len(outcomes)
    return probability_of_heads

# Problem 2: Probability of rolling a 4 on a six-sided die
def die_roll_probability():
    sides = [1, 2, 3, 4, 5, 6]
    probability_of_4 = sides.count(4) / len(sides)
    return probability_of_4

# Running the functions and printing the results
prob_heads = coin_flip_probability()
prob_4 = die_roll_probability()

print(f"Probability of getting heads in a coin flip: {prob_heads}")
print(f"Probability of rolling a 4 on a six-sided die: {prob_4}")
```

```python
#Normal Distribution
#The Normal distribution, also known as the Gaussian distribution, is a continuous probability distribution.
#It has a bell-shaped curve, symmetric around the mean (μ).
#It's characterized by two parameters: the mean (μ) and the standard deviation (σ).
#Mean (μ): The center of the distribution.
#Standard Deviation (σ): Controls the spread of the distribution.
#Code Example:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu = 0    # Mean
sigma = 1 # Standard Deviation

# Generate data points
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Normal Distribution\nμ={mu}, σ={sigma}')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
