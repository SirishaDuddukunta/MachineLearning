## Eigenvectors, Eigenvalues
Eigenvectors:
Eigenvectors are special vectors that, when a linear transformation is applied to them, do not change their direction. They may get scaled (stretched or compressed) but they do not rotate or change direction.

Example: Imagine pushing a book across a table. If you push it in certain directions, it just slides in that direction without flipping or rotating. These directions are like eigenvectors—they are directions that remain unchanged except for their length.

Eigenvalues:
What They Are: Eigenvalues are the factors by which the eigenvectors are scaled during the transformation. They tell you how much the eigenvector is stretched or compressed.

Example: If you push the book and it moves twice as far as it would normally move, the "2" in this case is like an eigenvalue. It represents the amount by which the eigenvector is stretched.

Mathematical Definition:
In a mathematical sense, if you have a matrix 
A×v=λ×v
where:
A is the matrix representing the transformation,
v is the eigenvector,
λ (lambda) is the eigenvalue.
This equation means that applying the matrix 𝐴
A to the vector 
v only changes the vector's magnitude (by λ) but not its direction.

Eigenvalues and Eigenvectors are the scalar and vector quantities associated with matrices used for linear transformations. The vector that only changes by a scalar factor after applying a transformation is called an eigenvector, and the scalar value attached to the eigenvector is called the eigenvalue.

Eigenvectors are vectors that are associated with a matrix such that when the matrix multiplies the eigenvector, the resulting vector is a scalar multiple of the eigenvector. Eigenvectors are also called characteristic vectors and can only be found for square matrices. They are very useful in solving various problems involving matrices and differential equations

(A-λI)v = 0 where A is given matrix,λ is scalar value,I is identity matrix and 0 is null matrix

```python
import numpy as np
# Example matrix
A = np.array([[4, 2],
              [1, 3]])
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Display results
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
B = np.array([[6, 2, 1],
              [2, 3, 1],
              [1, 1, 1]])

eigenvalues_B, eigenvectors_B = np.linalg.eig(B)
print("Eigenvalues of B:", eigenvalues_B)
print("Eigenvectors of B:\n", eigenvectors_B)
```

Function Behavior
When you call np.linalg.eig(B), the function performs a series of computations internally:

Eigenvalues: The function first computes the eigenvalues of the matrix 
𝐵
B. These are the roots of the characteristic equation 
det(𝐵−𝜆𝐼)=0
det(B−λI)=0, where λ represents the eigenvalues.

Eigenvectors: After finding the eigenvalues, the function computes the corresponding eigenvectors. These are the non-zero vectors 
v that satisfy 𝐵𝑣=𝜆𝑣
Bv=λv for each eigenvalue λ.
The function doesn’t rely on what you name these variables—it only returns the outputs in a fixed order:

 1.First, the eigenvalues.
 2.Then, the eigenvectors.

Problem 2: Diagonalization of a Matrix

Given a matrix:

𝐶=(3 1 0 2)
Find its eigenvalues and eigenvectors, then verify if the matrix can be diagonalized.

```python
C = np.array([[3, 1],
              [0, 2]])

eigenvalues_C, eigenvectors_C = np.linalg.eig(C) #C−λI, (C−λI)v=0.
D = np.diag(eigenvalues_C) #Diagonal Matrix Creation:creates a diagonal matrix D using the eigenvalues.
P = eigenvectors_C #Eigenvector Matrix:Each column of P corresponds to an eigenvector of the matrix C.
P_inv = np.linalg.inv(P) #We compute inverse of eigenvector matrixP,This is necessary for the diagonalization process.

# Verify diagonalization
diagonalized_C = P @ D @ P_inv #reconstructs C from its eigenvalues and eigenvectors,C=PDP−1.
print("Eigenvalues of C:", eigenvalues_C)
print("Eigenvectors of C:\n", eigenvectors_C)
print("Diagonalized C:\n", diagonalized_C)

```

## Singular Value Decomposition
Singular vectors are mathematical constructs that arise from the Singular Value Decomposition (SVD) of a matrix. SVD is a fundamental technique in linear algebra that decomposes a matrix into three other matrices, revealing important properties and structures within the original matrix.

Right Singular Vectors:
The right singular vectors of a matrix A are the columns of the matrix V from the SVD. If
A=UΣVT,
These vectors are the eigenvectors of the matrix ATA, and they represent directions in the feature space associated with A.
 
Left Singular Vectors:
The left singular vectors of a matrix A are the columns of the matrix U in the SVD. In the same equation,
The left singular vectors are the columns of U.
These vectors are the eigenvectors of the matrix AAT and represent directions in the original space (of the rows of A). 

Formula: 
For a given matrix A of size m*n,the SVD can be expressed as:

A = U ΣVT
where:
U is an m*m orthogonal matrix containing the left singular vectors.
Σ is an m×n diagonal matrix containing the singular values (non-negative values).
VT is the transpose of an n×n orthogonal matrix containing the right singular vectors.


PROPERTIES:
1.Orthogonal:Vector is called Orthogonal which has a dot product of zero.
2.Number of singular vectors
Norm:length of 1.

What is Singular Value Decomposition (SVD)?
Singular Value Decomposition, or SVD, is like taking a big, messy pile of LEGO blocks (which represents a matrix) and organizing them into neat, easy-to-understand sections. These sections make it much simpler to see what kind of pieces you have and how they fit together.

Breaking Down the Big Box (Matrix)
When you do SVD, you're breaking down the big box into three smaller boxes:

Box 1 (U): This box holds instructions on how to arrange the rows of your LEGO blocks.
Box 2 (Σ - Sigma): This box tells you the most important pieces to focus on. It sorts the blocks by importance, with the biggest, most useful pieces at the front.
Box 3 (V^T): This box shows you how to arrange the columns of your LEGO blocks.

Singular Value Decomposition (SVD) is a mathematical technique used to decompose a matrix into three other matrices. It's a powerful tool in linear algebra and is widely used in various applications, including machine learning, signal processing, and statistics.

The SVD of a matrix 
A is defined as:

𝐴=𝑈Σ𝑉 transpose 𝑇

Where:
A: The original matrix, which can be any real or complex matrix of size m×n.
U: An m×m orthogonal matrix whose columns are called the left singular vectors of A.
Σ: A diagonal matrix of size m×n with non-negative real numbers (called singular values) on the diagonal and zeros elsewhere.
 
T : The transpose of an n×n orthogonal matrix 

V, whose columns are called the right singular vectors of A.

```python
#Implementation
# create matrix
import numpy as np
from scipy.linalg import svd
A = np.array([[3, 2, 2],
              [2, 3, -2]])
print("Original Matrix A:\n", A)
```

```python
#Perform SVD using NumPy - np.linalg.svd function  is used to perform SVD. 
l,sigma,rt = np.linalg.svd(A)

print("l,the orthogonal matrix that contains the left singular vectors of A matrix:\n", l)
print("sigma,1D array holding the singular values of A, which are non-negative and arranged in descending order:\n",sigma)
print("transpose of an orthogonal matrix containing the right singular vectors of A:\n", rt)
```

```python
#Convert Sigma to a Diagonal Matrix
#The sigma returned by NumPy is a 1D array of singular values. To get the full diagonal matrix Σ, 
# you need to convert it to a diagonal matrix.
Sigma = np.zeros((A.shape[0], A.shape[1])) #1.Creating a Zero Matrix -The shape of Sigma is 
#  (number of rows in A, number of columns in A)
Sigma[:len(sigma), :len(sigma)] = np.diag(sigma) #2.len(sigma) provides the number of singular values calculated from the SVD.
#np.diag(sigma) creates a diagonal matrix where the singular values (contained in sigma) are placed on the diagonal.
#The notation Sigma[:len(sigma), :len(sigma)] ensures that only the top-left square portion of Sigma (which corresponds to the 
#dimensions of the diagonal matrix) is filled with the singular values. This effectively places the singular values along the
#diagonal of Sigma, while all other entries remain zero.

print("Sigma matrix:\n", Sigma)
```

```python
# Verify the Decomposition
#You can verify the decomposition by reconstructing the original matrix A using U,Σ and VT.
A_reconstructed = np.dot(l, np.dot(Sigma, rt)) #l is matrix of left singular vectors obtained from SVD,
# Sigma is diagonal matrix of singular values, np.dot(Sigma, rt)) performs matrix multipication,multiplies 
# the diagonal matrix of singular values by the transposed right singular vector matrix.
print("Reconstructed Matrix A:\n", A_reconstructed)
```

```python
#Perform SVD using SciPy 
#SciPy provides a similar svd function that can be used as well.
l_scipy, sigma_scipy, rt_scipy = svd(A)

print("L (SciPy):\n", U_scipy)
print("Sigma (SciPy):\n", sigma_scipy)
print("r^T (SciPy):\n", rt_scipy)
```

## SVD Applications: Image Compression Using SVD
This code demonstrates how to perform image compression using SVD by keeping only the most significant singular values, resulting in a compressed approximation of the original image.

Steps:
1.Convert an image into a grayscale matrix.
2.Perform SVD on this matrix.
3.Retain only the top k singular values and reconstruct the image.
4.Display the original and compressed images.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from skimage import data, color #data and color from skimage are used to
#access sample images and handle color conversions.
from skimage.transform import resize

# Load an image and convert it to grayscale
image = color.rgb2gray(data.astronaut()) #The astronaut image from skimage is 
#loaded and converted to grayscale using color.rgb2gray
image_resized = resize(image, (256, 256)) # resized to 256x256 pixels 

# Perform SVD
U, sigma, VT = svd(image_resized)

# Retain only the top 50 singular values
k = 50 #we will keep the top 50 singular values for reconstruction.
Sigma_k = np.zeros((image_resized.shape[0], image_resized.shape[1])) #A new diagonal matrix Sigma_k 
#of zeros is created, and the top 50 singular values are placed on its diagonal.
Sigma_k[:k, :k] = np.diag(sigma[:k]) #sigma[:k] takes the first k singular values from the sigma array,
#which represent the most significant components of the image and   np.diag(sigma[:k]) creates a diagonal
#matrix using these top k singular values.  Sigma_k[:k, :k] = ... assigns this diagonal matrix to the
#top-left k x k portion of the Sigma_k matrix.

# Reconstruct the image
image_compressed = np.dot(U, np.dot(Sigma_k, VT))

# Plot original and compressed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_resized, cmap='gray')
ax[0].set_title("Original Image")
ax[0].axis('off')

ax[1].imshow(image_compressed, cmap='gray')
ax[1].set_title(f"Compressed Image (k={k})")
ax[1].axis('off')

plt.show()
```
